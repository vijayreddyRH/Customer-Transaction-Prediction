#### Predictive Model Building

```python
# Feature & Target Separation
X = data.drop(['ID_code', 'target'], axis=1)
y = data['target']
```

```python
# Print the Feature
X
```

```python
# Print the Target
y
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
scaler
```

## Model Training

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_scaled, y_train)

y_pred_lr = lr.predict(X_test_scaled)
lr
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
dt
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
rf
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train, y_train)

y_pred_gb = gb.predict(X_test)
gb
```

## Model Evaluation

### Evaluation Function

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def evaluate(y_test, y_pred, model):
    return {
        "Model": model,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_pred)
    }
```

### Model Comparison Table

```python
results = []

results.append(evaluate(y_test, y_pred_lr, "Logistic Regression"))
results.append(evaluate(y_test, y_pred_dt, "Decision Tree"))
results.append(evaluate(y_test, y_pred_rf, "Random Forest"))
results.append(evaluate(y_test, y_pred_gb, "Gradient Boosting"))

results_df = pd.DataFrame(results)
results_df
```

# Report on Challenges Faced and Techniques Used
---
## Challenge 1: Anonymized Feature Names
**Problem**
* The dataset consists of 200 features whose meanings are not provided. Due to anonymization, it was not possible to understand feature importance from a domain perspective or perform traditional exploratory data analysis.

**Technique Used**
* Exploratory Data Analysis (EDA) based on feature interpretation was skipped as per project instruction. Instead, model-based analysis was used, focusing on evaluation metrics such as Accuracy, Precision, Recall, F1-Score, and ROC-AUC.
* Feature-based exploratory analysis was skipped as per project instruction. Instead, multiple machine learning models were trained and evaluated using performance metrics.

**Reason**
* When feature interpretation is not possible, model performance metrics provide a reliable way to assess the predictive capability of the models.

---

## Challenge 2: High Dimensional Data (200 Features)
**Problem**
* The dataset contains a high number of features (200), which increases model complexity and the risk of overfitting, especially for tree-based models.

**Technique Used**
* Multiple algorithms were tested, including Logistic Regression, Decision Tree, Random Forest, and Gradient Boosting. Model comparison was performed using consistent evaluation metrics.

**Reason**
* Testing multiple models helps identify which algorithm performs best when handling high-dimensional anonymized data.

---

## Challenge 3: Class Imbalance in Target Variable
**Problem**
* The target variable is imbalanced, with significantly fewer customers performing transactions. This caused some models to predict mostly the majority class.
* This is evident in the results where Random Forest and Gradient Boosting showed high accuracy but extremely low recall values.

**Technique Used**
* Stratified train-test split was applied to maintain class distribution. Model evaluation focused on Precision, Recall, F1-Score, and ROC-AUC rather than accuracy alone.

**Reason**
* Accuracy can be misleading in imbalanced datasets. Recall and F1-Score better represent the model’s ability to identify customers who will make transactions.

---

## Challenge 4: Poor Minority Class Detection by Ensemble Models Problem
**Problem**
* Random Forest and Gradient Boosting models achieved high accuracy but showed very low recall, indicating failure to identify customers who will perform transactions.

**Technique Used**
* Model comparison was performed to analyze all evaluation metrics. Logistic Regression was selected as the best model due to better balance between precision and recall.

**Reason**
* A model with balanced performance across metrics is more reliable than a model with high accuracy but poor recall in real-world banking applications.

---

## Challenge 5: Model Selection for Production
**Problem**
* Different models showed varying performance across metrics, making model selection challenging.

**Technique Used**
* A model comparison report was created using Accuracy, Precision, Recall, F1-Score, and ROC-AUC. Logistic Regression achieved the highest ROC-AUC (0.6257) and F1-Score (0.3811)

**Reason**
* ROC-AUC and F1-Score are important metrics for imbalanced classification problems, making Logistic Regression the most suitable model for deployment.

---

# Final Conclusion
Despite challenges such as anonymized features, high dimensionality, class imbalance, and inconsistent model behavior, a reliable predictive model was developed. Based on evaluation results, Logistic Regression was selected as the best model due to its superior ability to identify customers likely to make future transactions.
